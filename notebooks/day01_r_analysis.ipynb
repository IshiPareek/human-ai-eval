Response length and topic domain together explain only 3.4% of win rate variation for models with a win rate 75-50%, a 4.5% for models with a win rate between 50-75% and a 0.3% for models with a win rate between 0-25%.

Suggesting human preference is driven by factors beyond surface-level response characteristics like length and subject matter.

```r
# Calculate Linear Reg between Structure, Tone, Writing Style and Wiin Rate 

library(readr)
library(dplyr)

set2 <- read_csv('/Users/ishikapareek/Desktop/CURRENT PROJECTS/human-ai-eval/data/set2_topics.csv')
set2$response_length <- nchar(set2$response)

lm.fit2 <- lm(win_rate ~ Structure + Tone + Style, data = set2)
summary(lm.fit2)

par(bg = "white")
plot(lm.fit2, lwd = 2, pch = 16, bg='white')
```

Human evaluators prefer AI responses that are instructional and use markdown formatting over responses that are conversational or use bullet points. Humor, while rare, positively correlates with winning.

```r
# Let's see the change in win rate for writing style per domain 

library(readr)
library(dplyr)

set2 <- read_csv('/Users/ishikapareek/Desktop/CURRENT PROJECTS/human-ai-eval/data/set2_topics.csv')

top5_domains <- c('General Knowledge')  #Programming / Coding', 'General Explainer', 'Math / Logic', 'Science')

for (domain in top5_domains) {
    subset_data <- set2[set2$`topic/domain` == domain, ]
    fit <- lm(win_rate ~ Structure + Tone + Style, data = subset_data)
    cat("\n====", domain, "====\n")
    cat("Rows:", nrow(subset_data), "\n")
    cat("R-squared:", round(summary(fit)$r.squared, 4), "\n")
    print(round(summary(fit)$coefficients, 4))
}

par(bg = "white")
plot(lm.fit2, lwd = 2, pch = 16, bg='white')
```
